## squeeze unsqueeze
*squeeze的用法主要就是对数据的维度进行压缩或者解压*
- torch.squeeze() 这个函数主要对数据的维度进行压缩，去掉维数为1的的维度，比如是一行或者一列这种
- torch.unsqueeze()这个函数主要是对数据维度进行扩充。给指定位置加上维数为一的维度

```python
# unsqueeze 新增一个组别
a.unsqueeze(0).shape
```

```python
a.unsqueeze(-1).shape
```

```python
#squeeze 维度删减
b=torch.rand(1,32,1,1)
b.shape
```

```python
b.squeeze().shape
```

```python
b.squeeze(0).shape
```

```python
print(b.squeeze(-1).shape)
print(b.squeeze(1).shape) # 1 的位置是32 不能挤压
```

# 激活函数与梯度

### torch.linspace的详细用法
torch.linspace(start, end, steps=100, out=None) → Tensor
参数：
- start (float) - 区间的起始点
- end (float) - 区间的终点
- steps (int) - 在start和end间生成的样本数
- out (Tensor, optional) - 结果张量

```python
a=torch.linspace(-100,100,10)
```

```python
a
```

```python
torch.sigmoid(a)
```

#### MSE


```python
# autograd.grad pytorch自动求导
x=torch.ones(1)
```

```python
w=torch.full([1],2.)
```

```python
mse=F.mse_loss(torch.ones(1),x*w)
```

```python
mse
```

```python
torch.autograd.grad(mse,[w])
```

```python
w.requires_grad_() # grad后的_表示对w信息进行更新  w声明时即指定需要求导
```

```python
mse=F.mse_loss(torch.ones(1),x*w)
```

```python
torch.autograd.grad(mse,[w])
```

## softmax

和为1  
大的扩的更大，小的差距变小




```python
a=torch.rand(3)
```

```python
a.requires_grad_()
```

```python
p=F.softmax(a,dim=0)
```

```python
p.backward(torch.ones_like(a),retain_graph=True)
```

```python
torch.autograd.grad(p[1],[a],retain_graph=True)
```

```python
torch.autograd.grad(p[2],[a])
```
