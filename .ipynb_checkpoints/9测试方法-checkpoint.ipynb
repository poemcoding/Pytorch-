```python
import torch
import torch.nn as nn
from torch.nn import functional as F
```

```python
logits=torch.rand(4,10)
pred=F.softmax(logits,dim=1)
pred.shape
```

```python
pred_label=pred.argmax(dim=1)
pred_label
```

```python
logits.argmax(dim=1)
```

```python
label=torch.tensor([7,1,2,1])
correct=torch.eq(pred_label,label) #eq返回相同类型的Tensor，相等返回1，不等返回0
correct
```

```python
correct.sum().float().item()/4
```
