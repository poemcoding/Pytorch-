```python
import torch
import torch.nn as nn
from torch.nn import functional as F
```

```python
x=torch.randn(1,784)
x.shape
```

```python
# 通过高阶的API调用，省去了编写内部的很多步骤，参数顺序为input，output
layer1=nn.Linear(784,200)
layer2=nn.Linear(200,200)
layer3=nn.Linear(200,10)
```

```python
x=layer1(x)
x.shape
```

```python
x=layer2(x)
x.shape
```

```python
x=layer3(x)
x.shape
```

**relu**

```python
x=torch.randn(1,784)
x=layer1(x)
x=F.relu(x,inplace=True) # inplace为true，回收不用的Tensor空间
x.shape
```

```python
x=layer2(x)
x=F.relu(x,inplace=True) # inplace为true，回收不用的Tensor空间
x.shape
```

```python
x=layer3(x)
x=F.relu(x,inplace=True) # inplace为true，回收不用的Tensor空间
x.shape
```

```python
# 自己定义模型
class MLP(nn.Module):

    def __init__(self):
        super(MLP,self).__init__()

        self.model=nn.Sequential( # can add any model from nn.Model
            nn.Linear(784,200),
            nn.ReLU(inplace=True),
            nn.Linear(200,200),
            nn.ReLU(inplace=True),
            nn.Linear(200, 10),
            nn.ReLU(inplace=True),
        )

    def forward(self,x):
        x=self.model(x)

        return x
```

### nn.ReLU  vs  F.relu

写法不同；F.relu可自己添加定义项  
一般使用nn.ReLU
