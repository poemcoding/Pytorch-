 # Entropy
 entropy: 熵，不确定性，uncertainty
 
 是uncertainty的衡量度

```python
import torch
from torch.nn import functional as F
```

```python
a=torch.full([4],1/4.) #4个人均等概率中奖
a
```

```python
a*torch.log2(a)
```

```python
-(a*torch.log2(a)).sum() # 计算entropy
```

```python
a=torch.tensor([0.1,0.1,0.1,0.7])
```

```python
-(a*torch.log2(a)).sum()
```

```python
a=torch.tensor([0.001,0.001,0.001,0.998])
```

```python
-(a*torch.log2(a)).sum() #entropy越大不确定性越大，其中一个是0.998,大概率选择这个
```

## binary classification二分类

P()表示真实数据的概率，0/1  
Q()表示基于模型给出的概率

**例子**



如果新的领域，可以尝试MSE，可能会更好

```python
x=torch.randn(1,784)
```

```python
w=torch.randn(10,784)
logits=x@w.t()
logits.size()
```

```python
pred=F.softmax(logits,dim=1)
pred.size()
```

```python
pred_log=torch.log(pred)
pred_log
```

```python
F.cross_entropy(logits,torch.tensor([3])) # cross_entropy=softmax+log+null_loss
```

```python
F.nll_loss(pred_log,torch.tensor([3]))
```
