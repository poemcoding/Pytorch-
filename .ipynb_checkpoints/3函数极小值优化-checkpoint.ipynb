
函数极小值点

### Plot

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
def himmelblau(x):
    return (x[0]**2+x[1]-11)**2+(x[0]+x[1]**2-7)**2
x=np.arange(-6,6,0.1)
y=np.arange(-6,6,0.1)
print('x,y range:',x.shape,y.shape)
X,Y=np.meshgrid(x,y)
print('X,Y maps:',X.shape,Y.shape)
Z=himmelblau([X,Y])

fig=plt.figure('himmelblau')
ax=fig.gca(projection='3d')
ax.plot_surface(X,Y,Z)
ax.view_init(60,-30)
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

### 随机梯度下降  gradient descent

```python
# 对predict进行微分，不对error
x=torch.tensor([0.,0.],requires_grad=True)  # 不同初始化的值对应得到的结果不同，初始化很重要
optimizer=torch.optim.Adam([x],lr=1e-3)  # 优化器，学习率1e-3 自动完成x'=x-0.001Δx
for step in range(20000):
    
    pred=himmelblau(x)
    
    optimizer.zero_grad() #清零
    pred.backward()  #生成x,y梯度信息
    optimizer.step()  # 对x,y更新
    
    if step % 2000==0:
        print('step{}:x={},f(x)={}'.format(step,x.tolist(),pred.item()))
```

```python
x=torch.tensor([-4.,0.],requires_grad=True) 
optimizer=torch.optim.Adam([x],lr=1e-3)  # 优化器，学习率1e-3 自动完成x'=x-0.001Δx
for step in range(20000):
    
    pred=himmelblau(x)
    
    optimizer.zero_grad() #清零
    pred.backward()  #生成x,y梯度信息
    optimizer.step()  # 对x,y更新
    
    if step % 2000==0:
        print('step{}:x={},f(x)={}'.format(step,x.tolist(),pred.item()))
```
