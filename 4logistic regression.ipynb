**regression 和 classification的差异**

regression:  
目标让 pred=y 通过减小pred和y的差距  

classification：  
放大accuracy等指标  
通过对probability进行操作，Pr:real data  P$_\theta$:模型数据的P


- 第一种情况  
调节weight，预测值改变了（接近真实值），但是未大于0.5，未产生根本变化。gradient为0
- 第二种情况  
w值变化，0.499->0.501，改变较小，但accuracy改变大，会发生不连续的变化
