### regularization
cross-entropyloss 加上了新的项  $\theta$是网络参数，w1,b1,乘一个因子$\lambda$表示后项与前面的相关性。

通过设置参数，使网络退化成更少次方的网络结构

也叫做weight decay
#### regularization后的结果

对于数据集较少情况，防止将噪声纳入模型

L2更为常用
