# 感知机
## 多输出

```python
import torch
from torch.nn import functional as F
```

```python
x=torch.randn(1,10)
```

```python
w=torch.randn(2,10,requires_grad=True)
```

```python
o=torch.sigmoid(x@w.t())
```

```python
o.shape
```

```python
loss=F.mse_loss(torch.ones(1,2),o)
```

```python
loss
```

```python
loss.backward()
```

```python
w.grad
```
